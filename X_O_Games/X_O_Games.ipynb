# Крестики - нолики

```python
# Старт игры

def start():
    # Знакомим с правилами
    print ("Предлагаем вам сыграть в игру крестики-нолики")
    print ("Для участив игре вам потребуется:")
    print ("1.Ввести ваши имена")
    print (" ")
    print ("2.Вводить координаты")
    print ("вашего хода в формате:")
    print (" ")
    print ("номер стр. номер столбца.")
    
    # Определяем первую двойку игроков
    player_1, player_2 = [input("Введите имея игрока ") for i in range(2)]
    
    return player_1, player_2

# Создаем игровое поле

cells = [[" "]*3 for i in range (3)] # Создаем поле для игры

def axes(): # Функция вывода поля
    print (f" 0 1 2") # Ось х
    for i in range (3):
        row = " ".join(cells[i])
        print (f"{i}{row}") # Ось y, печатаем i символ в row строке
        
# Определяем ходы игроков
        
def enter_cords(): # Функция ввода координат + проверка
    while True:
        cords = input ("Введите координаты вашего хода:").split() # Заправшиваем координаты
        
        if len(cords) != 2: # Проверяем кол-во
            print ("Координат должно быть 2!")
            continue
        x, y = cords # Разделяем данные строки по координатам
        
        if not(x.isdigit()) or not(y.isdigit()): # Проверяем что данные это числа
            print ("Введенные данные не являются числами!")
            continue
        x, y = int(x), int(y) # Преобразовываем из типа Str в тип Int
            
        if 0 > x or x > 2 or 0 > y or y > 2: # Проверка на соответствие диапазону поля
            print ("Полученные координаты не верны!")
            continue
            
        if cells [x][y] != " ": # Проверяем занята ли область
            print ("Клетка занята!")
            continue
            
        return x, y  
    
# Определяем ходы игроков

def ref():
    win_comb = [((0, 0),(0, 1),(0,2)), ((1, 0),(1, 1),(1,2)), ((2, 0),(2, 1),(2,2)),
                ((0, 2),(1, 1),(2,0)), ((0, 0),(1, 1),(2,2)), ((0, 0),(1, 0),(2,0)),
                ((0, 1),(1, 1),(2,1)), ((0, 2),(1, 2),(2,2))] # Условия победы
    
    for cords in win_comb: # Сравниваем игровое поле с выигрышными комбинациями
        a = cords[0]
        b = cords[1]
        c = cords[2]
        
        if cells [a[0]][a[1]] == cells [b[0]][b[1]] == cells [c[0]][c[1]] == "X": # Кто бобедит если собираем х
            print("Победа игрока", player_1)
            return True
        elif cells [a[0]][a[1]] == cells [b[0]][b[1]] == cells [c[0]][c[1]] == "O": # Кто бобедит если собираем о
            print("Победа игрока", player_2)
            return True
    return False

player_1,player_2 = start()
counter = 0 # Счетчик ходов
while True:
    counter +=1
    
    axes() # Выводим оси для игры
    
    if counter % 2 == 1: # Условия для опеределения очередности ходов поп ринципу чет /нечет
        print (player_1,"!, Поставьте крестик")
    else:
        print (player_2,"!, Поставьте нолик")  
        
    x, y = enter_cords()
        
    if counter % 2 == 1: # Условия для выбора игрового символа
        cells[x][y] = "X"
    else:
        cells[x][y] = "O"
        
    if ref():
        break
    
    if counter == 9:
        print ("Ничья")
```
